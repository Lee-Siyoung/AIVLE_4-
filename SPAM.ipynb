```python
def clean_documents(documents):
    #텍스트 정제 (HTML 태그 제거)
    for i, document in enumerate(documents):
        document = BeautifulSoup(document, 'html.parser').text 
        documents[i] = document

    #텍스트 정제 (특수기호 제거)
    for i, document in enumerate(documents):
        document = re.sub(r'[^ ㄱ-ㅎㅏ-ㅣ가-힣]', ' ', document) #특수기호 제거, 정규 표현식 
        documents[i] = document

    return documents
```
